# Operations

We can perform a multitude of operations using arrays.

```python
import numpy as np
```

```python
# Creating a simple array
arr = np.arange(0,11)
```

# max,min,argmax,argmin

These are useful methods for finding max or min values. Or to find their index locations using argmin or argmax.

```python
arr.max()
```

```python
arr.min()
```

```python
arr.argmax()
```

```python
arr.argmin()
```

# Arithmetic

```python
# Sum
print(arr + arr)

# Multiply
print(arr * arr)

# Substract
print(arr - arr)
```

```python
# Warning on division by zero, but not an error!
# Just replaced with nan
arr/arr
```

```python
# Also warning, but not an error instead infinity
1/arr
```

```python
arr**3
```

# Universal Array Functions

Numpy comes with many universal array functions, which are essentially just mathematical operations you can use to perform the operation across the array. 

[Documentation](https://numpy.org/doc/stable/reference/ufuncs.html)

```python
#Taking Square Roots
np.sqrt(arr)
```

```python
#Calcualting exponential (e^)
np.exp(arr)
```

```python
#Get maximum of an array
#same as arr.max()
np.max(arr) 
```

```python
np.sin(arr)
```

```python
np.log(arr)
```
