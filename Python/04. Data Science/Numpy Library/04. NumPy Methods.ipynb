# NumPy Methods

As with lists, we are able to use various methods on arrays. We can sort, delete or append items to an array. We can even merge two arrays. 

```python
import numpy as np
```

```python
# Lets create two arrays

arr = np.arange(25)
arr2 = arr.reshape(5,5)
```

```python
arr
```

```python
arr2
```

# Append

`np.append` will add an item to an array at the end of it.

```python
arr = np.append(arr, 25)
arr
```

```python
# Make sure to put .reshape() at the end of the line, otherwise it will make it flat

arr2 = np.append(arr2, [25,26,27,28,29]).reshape(6,5)
arr2
```

# Sort

`np.sort` will arrange an array from small to large, as it would in a list.

```python
# Creating a new array for sorting

arrx = np.array([1,2,5,8,124,5688,21,79890,3,4])
arrx
```

```python
np.sort(arrx)
```

# Delete

Using `np.delete`, we can delete items that are in an certain index position.

```python
arrx = np.delete(arrx, 5)
arrx
```

# Concatenate

We can concatenate two arrays using `np.concatenate()` into a single array.

```python
# Creating two 1d arrays

a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
```

```python
# Concatenation

np.concatenate((a, b))
```

```python
# Creating 2 arrays, including one 2d

x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6]])

# Concatenation
np.concatenate((x, y), axis=0)
```
