# Groupby

The groupby method allows you to group rows of data together and call aggregate functions

```python
import pandas as pd
```

```python
# Createing a dataframe

df = pd.DataFrame({'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]})
df
```

The `df.groupby()` method allows us to group rows together based off of a column name. For instance let's group based off of Company. This will create a DataFrameGroupBy object.

```python
# Grouping the data based on the Company

df.groupby('Company')
```

```python
# Saving the result to a variable

by_comp = df.groupby('Company')
```

It order to display our result, we have to use an aggregate function so that python knows how to deal with the data.

```python
# Using mean to get averages

by_comp.mean()
```

```python
# We can do everything in one line

df.groupby('Company').mean()
```

```python
# Getting the average for FB

df.groupby('Company').mean().loc["FB"]
```

```python
# Getting the maximum
# In this case, we will also get the person latest on the alphabet

df.groupby('Company').max()
```

```python
# Standard deviation

df.groupby('Company').std()
```

```python
# Getting a count of individual items

df.groupby('Company').count()
```

Using `describe()` will get more statistical information regarding our DataFrame. 

```python
df.groupby('Company').describe()
```

```python
# Transposing our described DataFrame

df.groupby('Company').describe().transpose()
```

```python
# Described data for GOOG

df.groupby('Company').describe().transpose()['GOOG']
```
