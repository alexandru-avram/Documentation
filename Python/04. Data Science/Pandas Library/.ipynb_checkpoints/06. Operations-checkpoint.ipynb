# Operations

There are lots of operations with pandas that will be really useful to you

```python
import pandas as pd
```

```python
# Creating a simple DataFrame

df = pd.DataFrame({'col1':[1,2,3,4],'col2':[444,555,666,444],'col3':['abc','def','ghi','xyz']})
df.head()
```

# Info on Unique Values

```python
# Finding all the unique values in a column

df['col2'].unique()
```

```python
# Finding all the unique values in a row

df.loc[0].unique()
```

```python
# Number of unique values

df['col2'].nunique()
```

```python
# Counting unique values

df['col2'].value_counts()
```

# Applying Functions

#### apply(function)

```python
# Creating a simple function

def times2(x):
    return x*2
```

```python
# Use apply to initiate the function

df['col1'].apply(times2)
```

```python
# Works with built-in functions

df['col3'].apply(len)
```

```python
# We can use lambda functions with apply

df['col2'].apply(lambda x: x**2)
```

# Sorting and Ordering a DataFrame

#### df.sort_value(by)

We can sort and order our DataFrame using `df.sort_values`. Remember that `inplace = False` by default

```python
df
```

```python
# Note that the index values remains attached to the rows

df.sort_values(by='col2') 
```

```python
# Use reset_index

df.sort_values(by='col2').reset_index()
```

# Pivot Table

#### df.pivot_table(values, index, columns)

The pivot table is an important tool when working with pandas. It is esentailly the same as pivot tables from Excel. 

```python
df = pd.DataFrame({'A':['foo','foo','foo','bar','bar','bar'],
     'B':['one','one','two','two','one','one'],
       'C':['x','y','x','y','x','y'],
       'D':[1,3,2,5,4,1]})

display(df)
```

```python
# Creating a pivot table based on the D column
# A and B are grouped in a multilevel index

df.pivot_table(values='D',index=['A', 'B'],columns=['C'])
```
