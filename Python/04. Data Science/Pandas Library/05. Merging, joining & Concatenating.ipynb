# Data Modeling

Data modeling is an important part of the data analysis process. Pandas is equiped with multiple tools to help python users not just in grouping DataFrames, but also merging, joining and concatenating multiple DataFrames.

```python
import pandas as pd
```

# Concatenation

#### pd.concat()

Concatenation basically glues together DataFrames. Keep in mind that dimensions should match along the axis you are concatenating on.

First, we will need to create multiple DataFrames to showcase this.

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
print("df1")
display(df1)

print("\ndf2")
display(df2)

print("\ndf3")
display(df3)
```

```python
# Concatenating all DataFrames

pd.concat([df1,df2,df3])
```

```python
# Concatenating all DataFrames on axis 1

pd.concat([df1,df2,df3], axis = 1)
```

# Merging

#### merge( dataframe1, dataframe2, how, on)

The merge function allows you to merge DataFrames together using a similar logic as merging SQL Tables together.

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})

print("left")
display(left)

print("\nright")
display(right)
```

```python
pd.merge(left, right, on=['key1', 'key2'])
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left, right, on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

# Joining

Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame. 

The difference between merging and joining is that the key is on the index.

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

display(left)

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])

display(right)
```

```python
left.join(right)
```

Just like `merge()`, we can specify the `how=`.

```python
left.join(right, how='outer')
```
