# Types of Numbers

Unlike SQL which can get you confused with various data types, Python uses mainly two numerical data types:

* **Integers** - whole numbers, like 2,3,4,5 or -2
* **Floating** - numbers that have a decimal point, like: 2.3, 2.902 or even 2.0

# Simple Arithmetic

```python
#The Basics: Addition, Substraction, Multiplication and Division

print(6 + 2)
print(6 - 2)
print(6 * 2)
print(6 / 3)
```

```python
#The Basics part 2: using paranthesis

(6 + 2) * (2 * 3)
```

```python
#The Basics part 3:

#Power
print(6**2)

#Square roots
print(6**0.5)

#Floor division
print(13//2)

#Division remainder
print(13%2)
```

## Calculating Net Income

*** based on the Romanian Tax Code

```python
gross_income = 100
pensions = 0.25
healthcare = 0.1
income_tax = 0.1


net_income = (gross_income - (gross_income * (pensions + healthcare))) * (1 - income_tax)

net_income
```

## Numeric Functions

There are a number of simple numeric built-in functions that can be employed easily.

```python
# MIN - returns the lowest value from a list

print(min(1,2,3,4,5,-1))
```

```python
# MAX - returns the highest value from a list

print(max(1,2,3,4,5,-1))
```

```python
#ABS - returns the absolute value, when the number is less then 0

print(abs(-145))
```

```python
# ROUND - roads a number to a certain number of decimals, or to the closest integer value

print(round(145.123456789,2))
print(round(145.9))
```

```python
#SUM - adds the number for a list

print(sum([1,2,3,4,5,6,7]))
```
